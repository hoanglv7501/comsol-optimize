```python
threshold=0.4
j=0
def value(i):
    return onlyint[onlyint[a]==alpha].values[i,2]

alpha=listalpha[4]
for alpha in listalpha[:]:
    freq=[]
    R=[]
    j+=1
    for i in range(1,len(onlyint[onlyint[a]==alpha].values[:,1])-1):
        if (abs(value(i+1)+value(i-1)-2*value(i))>threshold*(abs(value(i)-value(i-1)))): 
            freq.append(onlyint[onlyint[a]==alpha].values[i,1])
            R.append(onlyint[onlyint[a]==alpha].values[i,2])

    frange=[]
    Rrange=[]
    frange.append(freq[0])
    Rrange.append(R[0])
    for i in range(len(freq)-1):
        if freq[i+1]-freq[i]>10: 
            frange.append(freq[i])
            Rrange.append(R[i])
            frange.append(freq[i+1])
            Rrange.append(R[i+1])
    frange.append(freq[-1])
    Rrange.append(R[-1])
    print(frange)
    print(f"alpha = {alpha} number {alpha*180/math.pi}")
    # plt.scatter(new[new[a]==alpha].values[:,1],new[new[a]==alpha].values[:,2])
    plt.figure(figsize=(20,5))
    plt.plot(onlyint[onlyint[a]==alpha].values[:,1],onlyint[onlyint[a]==alpha].values[:,2])
    plt.scatter(onlyint[onlyint[a]==alpha].values[:,1],onlyint[onlyint[a]==alpha].values[:,2])
    plt.scatter(freq,R,color='green')
    plt.scatter(frange,Rrange,color='red')
    plt.show()


# for alpha in listalpha[:3]:
#     freq=[]
#     R=[]
#     for i in range(1,len(onlyint[onlyint[a]==alpha].values[:,1])-1):
#         if (value(i)>value(i+1) and value(i)>value(i-1)) or (value(i)<value(i+1) and value(i)<value(i-1)):
#             freq.append(onlyint[onlyint[a]==alpha].values[i,1])
#             freq.append(onlyint[onlyint[a]==alpha].values[i-1,1])
#             freq.append(onlyint[onlyint[a]==alpha].values[i+1,1])
#             R.append(onlyint[onlyint[a]==alpha].values[i,2])
#             R.append(onlyint[onlyint[a]==alpha].values[i-1,2])
#             R.append(onlyint[onlyint[a]==alpha].values[i+1,2])
    
#     frange=[]
#     Rrange=[]
#     frange.append(freq[0])
#     Rrange.append(R[0])
#     for i in range(len(freq)-1):
#         if freq[i+1]-freq[i]>10: 
#             frange.append(freq[i])
#             Rrange.append(R[i])
#             frange.append(freq[i+1])
#             Rrange.append(R[i+1])
#     frange.append(freq[-1])
#     Rrange.append(R[-1])
#     print(frange)
#     print(f"alpha = {alpha} number {alpha*180/math.pi}")
#     # plt.scatter(new[new[a]==alpha].values[:,1],new[new[a]==alpha].values[:,2])
#     plt.figure(figsize=(20,5))
#     plt.plot(onlyint[onlyint[a]==alpha].values[:,1],onlyint[onlyint[a]==alpha].values[:,2])
#     # plt.scatter(onlyint[onlyint[a]==alpha].values[:,1],onlyint[onlyint[a]==alpha].values[:,2])
#     plt.scatter(freq,R,color='green')
#     plt.scatter(frange,Rrange,color='red')
#     plt.show()
```
