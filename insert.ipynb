```python
import pandas as pd
import numpy as np
from preprocess import getData,Transform2D
from math import pi
```

```python
rough=getData("rough.csv",convert_angle='rad2deg')
deep=getData("deep.csv")


a=list(rough)[0]
f=list(rough)[-2]
R=list(rough)[-1]

rough[a]=np.round(rough[a],3)
deep[a]=np.round(deep[a],3)
deep=deep.drop(columns=[list(deep)[1]])
```

```python

list_a=pd.unique(rough[a])
list_f=pd.unique(rough[f])
for alpha in list_a:
    rough=rough.append(pd.DataFrame({a:[alpha],f:[650],R:[rough[rough[a]==alpha][R].values[-1]]}))
```

```python
step_int=list_f[1]-list_f[0]
step_new=step_int/3

finalsweep=pd.DataFrame()

for alpha in list_a:
    # rough=rough.append(pd.DataFrame({a:[alpha],f:[600],R:[rough[rough[a]==alpha][R].values[-1]]}))
    for freq in np.linspace(500,650,101):
        if np.round(freq,5) in np.round(rough[rough[a]==alpha][f].values,5):
            finalsweep=finalsweep.append(rough[(rough[f]==freq)&(rough[a]==alpha)])
            
        elif np.round(freq,5) in np.round(deep[deep[a]==alpha][f].values,5):
            finalsweep=finalsweep.append(deep[(np.round(deep[f],5)==np.round(freq,5))&(deep[a]==alpha)])
        else:
            print(freq)
            nearf=rough[(rough[a]==alpha)&(rough[f]<freq)][f].values[-1]
            nextf=rough[(rough[a]==alpha)&(rough[f]>freq)][f].values[0]
            R1=rough[(rough[f]==nearf)&(rough[a]==alpha)][R].values[0]
            R2=rough[(rough[f]==nextf)&(rough[a]==alpha)][R].values[0]
            TotalR=(R2-R1)/step_int*(freq-nearf)+R1
            casealpha=pd.DataFrame({a:[alpha],f:[freq],R:[TotalR]})
            finalsweep=finalsweep.append(casealpha)
finalsweep.to_csv(r"new.csv",index=False)
```
